```sql
CREATE OR REPLACE TEMP VIEW v_driver_standings_2018
AS 
SELECT race_year, driver_name,team, total_points, wins, rank
FROM driver_standings
WHERE race_year = 2018
```

```sql
SELECT * FROM v_driver_standings_2018
```

```sql
CREATE OR REPLACE TEMP VIEW v_driver_standings_2020
AS 
SELECT race_year, driver_name,team, total_points, wins, rank
FROM driver_standings
WHERE race_year = 2020;
```

```sql
SELECT * FROM v_driver_standings_2020;
```

# Inner Join

```sql
SELECT *
FROM v_driver_standings_2018 AS d18
JOIN v_driver_standings_2020 AS d20
ON d18.driver_name = d20.driver_name
```

# LEFT JOIN

```sql
SELECT *
FROM v_driver_standings_2018 AS d18
LEFT JOIN v_driver_standings_2020 AS d20
ON d18.driver_name = d20.driver_name
```

# RIGHT JOIN

```sql
SELECT *
FROM v_driver_standings_2018 AS d18
RIGHT JOIN v_driver_standings_2020 AS d20
ON d18.driver_name = d20.driver_name
```

# FULL OUTER JOIN

```sql
SELECT *
FROM v_driver_standings_2018 AS d18
FULL JOIN v_driver_standings_2020 AS d20
ON d18.driver_name = d20.driver_name
```

# SEMI

```sql
SELECT *
FROM v_driver_standings_2018 AS d18
SEMI JOIN v_driver_standings_2020 AS d20
ON d18.driver_name = d20.driver_name
```

# ANTI

```sql
SELECT *
FROM v_driver_standings_2018 AS d18
ANTI JOIN v_driver_standings_2020 AS d20
ON d18.driver_name = d20.driver_name
```

# CROSS

```sql
SELECT *
FROM v_driver_standings_2018 AS d18
CROSS JOIN v_driver_standings_2020 AS d20
```
