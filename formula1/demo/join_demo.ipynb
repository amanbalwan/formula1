```python
circuits_df = spark.read.parquet(f'{processed_folder_path}/circuits')\
    .withColumnRenamed("name","circuit_name")
```

```python
display(circuits_df)
```

```python
races_df = spark.read.parquet(f'{processed_folder_path}/races').filter("race_year == 2019")\
    .withColumnRenamed("name","race_name")
```

```python
display(races_df)
```

### Inner Join

```python
#join
race_ciruits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "inner")\
    .select(circuits_df.circuit_name, circuits_df.location, circuits_df.country, races_df.race_name, races_df.round)
```

```python
display(race_ciruits_df)
```

### Outer joins

```python
circuits_df = spark.read.parquet(f'{processed_folder_path}/circuits')\
    .filter("circuit_id < 70")\
    .withColumnRenamed("name","circuit_name")
```

```python
display(circuits_df)
```

```python
#left outer join
race_ciruits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "left")\
    .select(circuits_df.circuit_name, circuits_df.location, circuits_df.country, races_df.race_name, races_df.round)
```

```python
display(race_ciruits_df)
```

```python
# right outer join
race_ciruits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "right")\
    .select(circuits_df.circuit_name, circuits_df.location, circuits_df.country, races_df.race_name, races_df.round)
```

```python
display(race_ciruits_df)
```

```python
#full outer join
race_ciruits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "full")\
    .select(circuits_df.circuit_name, circuits_df.location, circuits_df.country, races_df.race_name, races_df.round)
```

```python
display(race_ciruits_df)
```

### Semi Join

```python
race_ciruits_df = circuits_df.join(races_df, circuits_df.circuit_id == races_df.circuit_id, "semi")
```

```python
display(race_ciruits_df)
```

```python
race_ciruits_df = races_df.join(circuits_df , circuits_df.circuit_id == races_df.circuit_id, "anti")
```

```python
display(race_ciruits_df)
```

```python
race_ciruits_df = circuits_df.crossJoin(races_df)
```

```python
display(race_ciruits_df)
```
