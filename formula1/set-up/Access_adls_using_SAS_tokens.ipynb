# Access ADLS using SAS tokens
- 1. set auth type to sas
- 2. set SAS provider type to Fixed
- 3. conect using SAS token
- 4. access the table using abfs
- 5. print the content

```python
formula1dl_sas_token = dbutils.secrets.get(scope = 'formula1-scope',key='formula1dl-demo-account-SAS')
```

```python
spark.conf.set("fs.azure.account.auth.type.formula1adls.dfs.core.windows.net","SAS")
spark.conf.set("fs.azure.sas.token.provider.type.formula1adls.dfs.core.windows.net",
               "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.formula1adls.dfs.core.windows.net", formula1dl_sas_token)
```

```python
display(dbutils.fs.ls("abfss://demo@formula1adls.dfs.core.windows.net"))
```

```python
display(spark.read.csv("abfss://demo@formula1adls.dfs.core.windows.net/circuits.csv"))
```
