# Access Adls using access keys
- set up configuration using spark
- check content of files using databricks utils
- print contents


```python
formula1dl_account_key = dbutils.secrets.get(scope = 'formula1-scope',key='formula1dl-account-key')
```

```python
spark.conf.set(
    "fs.azure.account.key.formula1adls.dfs.core.windows.net",
    formula1dl_account_key
)
```

```python
fileinfo=dbutils.fs.ls("abfss://demo@formula1adls.dfs.core.windows.net")
```

```python
display(fileinfo)
```

```python
display(spark.read.csv("abfss://demo@formula1adls.dfs.core.windows.net/circuits.csv"))
```
