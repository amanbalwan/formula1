# Access ADLS using Service Principal/Applications
- 1. create a service principal in Azure Directory
- 2. get the client id , tenant id and Secret/password
- 3. connect using spark configs
- 4. assign the role to the datalake for the service principal

```python
client_id= dbutils.secrets.get(scope = 'formula1-scope',key = 'formula1-app-client-id')
tenant_id= dbutils.secrets.get(scope = 'formula1-scope',key = 'formula1-app-tenant-id')
client_secret= dbutils.secrets.get(scope = 'formula1-scope',key = 'formula1-app-client-secret')
```

```python
spark.conf.set("fs.azure.account.auth.type.formula1adls.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.formula1adls.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.formula1adls.dfs.core.windows.net", client_id)
spark.conf.set("fs.azure.account.oauth2.client.secret.formula1adls.dfs.core.windows.net", client_secret)
spark.conf.set("fs.azure.account.oauth2.client.endpoint.formula1adls.dfs.core.windows.net", f"https://login.microsoftonline.com/{tenant_id}/oauth2/token")
```

```python
display(dbutils.fs.ls("abfss://demo@formula1adls.dfs.core.windows.net"))
```
