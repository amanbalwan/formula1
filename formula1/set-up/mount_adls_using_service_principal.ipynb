# Mount ADLS using Service Principal
- 1. Get the client_id, tenant_id and client_secret from azure key vault
- 2. create a dictionary with spark configs
- 3. mount the adls to the workspace
- 4. analyse the data 
- 5. display mount and unmount

```python
client_id= dbutils.secrets.get(scope = 'formula1-scope',key = 'formula1-app-client-id')
tenant_id= dbutils.secrets.get(scope = 'formula1-scope',key = 'formula1-app-tenant-id')
client_secret= dbutils.secrets.get(scope = 'formula1-scope',key = 'formula1-app-client-secret')
```

```python
configs = {"fs.azure.account.auth.type": "OAuth",
          "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
          "fs.azure.account.oauth2.client.id": client_id,
          "fs.azure.account.oauth2.client.secret": client_secret,
          "fs.azure.account.oauth2.client.endpoint": f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"}
```

```python
dbutils.fs.mount(
  source = "abfss://demo@formula1adls.dfs.core.windows.net/",
  mount_point = "/mnt/formula1adls/demo",
  extra_configs = configs)
```

```python
display(dbutils.fs.ls("/mnt/formula1adls/demo"))
```

```python
display(spark.read.csv('dbfs:/mnt/formula1adls/demo/circuits.csv'))
```

```python
display(dbutils.fs.mounts())
```

```python
dbutils.fs.unmount('/mnt/formula1adls/demo')
```
